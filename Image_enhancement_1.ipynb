```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def enhance_grayscale(image_path, output_path):
    # Load the grayscale image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply histogram equalization to enhance contrast
    enhanced_img = cv2.equalizeHist(img)

    # Apply Gaussian blur to reduce noise
    enhanced_img = cv2.GaussianBlur(enhanced_img, (3, 3), 0)

    # Save the enhanced image
    cv2.imwrite(output_path, enhanced_img)

    #Read the enhanced image - This line was moved here and modified to read the saved output file
    enhanced_img = cv2.imread(output_path, cv2.IMREAD_GRAYSCALE)

    # Return the enhanced image so it's accessible outside the function
    return enhanced_img

# Example usage
input_image = "/content/mountains-lake.jpg"  # Replace with your image path
output_image = "enhanced.jpg"
#Assign the returned value from function to a variable
enhanced_img = enhance_grayscale(input_image, output_image)
plt.imshow(enhanced_img, cmap='gray')
plt.axis('off')
plt.show()
```
